```python
sexp = simplify_nq_example(json.loc[3])

print( f'question: {sexp["question_text"]}' )

annots = sexp['annotations']
print(annots)
for a in annots:
    long_s = a['long_answer']['start_token']
    long_e = a['long_answer']['end_token'] 

    print(f"Long: {' '.join( sexp['document_text'].split()[long_s:long_e] )}")
    if a['short_answers']:

        answers = []
        for rng in a['short_answers']:        
            short_s = rng['start_token']
            short_e = rng['end_token']
            answers.append( ' '.join( sexp['document_text'].split()[short_s:short_e] ))
        print(answers)

        print("\n =========== \n")
```
