```python
import matplotlib.pyplot as plt
import numpy as np
import re
import sys
sys.path.append('..')

import analysis_utils as au

accuracies = {}

names = ['adapters', 'prefix-tuning-30len', 'prefix-tuning-200len', 'lora',  'in-ctx', 'finetuning']

for name in names:
    stat_data = au.parse_log_file(name)

    accuracies[name] = stat_data

    au.plot(stat_data, name, show=False)
    
plt.show()
    
    
```

```python
for name in names:
    acc = np.asarray(accuracies[name])
    x = acc[:,0]
    y = acc[:,1]
    plt.plot(x, y, label=name, linewidth=2.0) 


plt.xlim(1)
plt.xlabel('epoch')
plt.ylabel('exact_match accuracy')
plt.legend()
plt.show()
```
